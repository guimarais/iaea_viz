```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv('files/001/001-fpos-50-20.xyz', skiprows=2, names=['atom', 'x', 'y', 'z'], delimiter=r'\s*', engine='python')
```

```python
df.head()
```

```python
dfs = df[0:10000]
```

```python
fig = plt.figure(figsize=(8,5), dpi=100)
ax = fig.add_subplot(111, projection='3d')

ax.scatter(dfs['x'], dfs['y'], dfs['z'], s=1, c='k', alpha=0.1)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
